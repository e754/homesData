```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/cityofBoston2024.csv')
```

```python
category_counts = df['CITY'].value_counts()

# Create the bar plot
plt.figure(figsize=(40, 10))
category_counts.plot(kind='bar', color='skyblue')
plt.title('Cities')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```

```python
category_counts = df['EXT_COND'].value_counts()

# Create the bar plot
plt.figure(figsize=(8, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.title('Category Distribution')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.show()
```
